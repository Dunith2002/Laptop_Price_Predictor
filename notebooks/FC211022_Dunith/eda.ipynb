### Check whether there are missing values in dataset

```python
df_new = df.drop(columns=['Unnamed: 0'])

df_new.isnull().sum()
```

```python
def null(df):
    h=df.isnull().sum()
    g=pd.DataFrame(data=h,columns=["null"])
    g=g[g['null']>0]
    plt.figure(figsize=(16, 8))
    plt.bar(g.index,g['null'])
    plt.title("Number of empty values ")
    plt.show()

null(df) 
```

```python

plt.figure(figsize=(16, 8))
sns.heatmap(df_new.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

* Battery Life has the most number of missing values.
* Need to handle it using mean/median Battery life by Brand.

#### Univarient Analysis

##### Price

```python
print("Price Unique Values:", df['Price'].nunique())
print("Missing Price values:", df['Price'].isna().sum())
print(df['Price'].value_counts())
```

```python

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

mean_price = df['Price'].mean()
sns.histplot(df['Price'], kde=True, ax=axes[0])
axes[0].set_title('Distribution of Laptop Prices')

sns.boxplot(data=df, y='Price', ax=axes[1])
axes[1].set_title('Laptop Pricing Spread')
axes[1].set_ylabel('Price')

axes[1].axhline(mean_price, color='red', linestyle=':', linewidth=1, label='Mean')
axes[1].legend()

plt.tight_layout()
plt.show()
```

* Distribution is right-skewed
* There are outliars

##### Box Plots for Categorical Variables against Price

```python
# Here I only consider categorical columns with <15 unique values
for col in df.select_dtypes(include='object').columns:
    if df[col].nunique() < 15:
        plt.figure(figsize=(16, 6))
        sns.boxplot(data=df, x=col, y='Price')
        plt.xticks(rotation=45)
        plt.title(f'Price vs {col}')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

```
