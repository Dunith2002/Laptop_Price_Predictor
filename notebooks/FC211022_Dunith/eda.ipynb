Check whether there are missing values in dataset

```python
df_new = df.drop(columns=['Unnamed: 0'])

df_new.isnull().sum()
```

```python

sns.heatmap(df_new.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

Visualizing dataset

```python
# Pricing Distribution
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of Laptop Prices')
plt.show()
```

```python
#boxplot for laptop price
sns.boxplot(data=df, y='Price')
plt.title('Laptop Pricing Spread')
plt.ylabel('Price')
plt.show()
```

Feature Visualization

Brand 

```python

plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='Brand', palette='Set1')
plt.xticks(rotation=45)
plt.title("Laptop Brands Count")
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.lineplot(data=df,x='Brand',y='Price')
plt.xticks(rotation=90)
plt.title('The relationship between Brand and Price ')
```

RAM

```python

plt.figure(figsize=(10, 6)) 
sns.boxplot(data=df, x='RAM', y='Price')
plt.title("Price by RAM Size")
plt.xticks(rotation=45)
plt.show()
```
