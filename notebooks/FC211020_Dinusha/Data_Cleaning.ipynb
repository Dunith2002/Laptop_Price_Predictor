#  **Data Preprocessing & Cleaning**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/workspaces/Laptop_Price_Predictor/data/raw/laptop_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum() #Check sum of null values
```

```python
df = df.drop_duplicates() #Remove duplicates columns
```

```python
df.drop(df.columns[0], axis=1, inplace=True) #Remove Unnamed column
```

Convert String Values to Numeric

```python
#Clean RAM_Expandable column
df['RAM_Expandable'] = df['RAM_Expandable'].str.rstrip('GB Expandable') # Remove all non-numeric characters
df['RAM_Expandable'] = df['RAM_Expandable'].replace('Not', '0').astype(int)
df.rename(columns={'RAM_Expandable': 'RAM_Expandable (GB)'}, inplace=True)
```

```python
#Clean RAM column
df['RAM'] = df['RAM'].str.extract('(\d+)').astype(int)
df.rename(columns={'RAM': 'RAM_Size (GB)'}, inplace=True)
```
